```python
import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("preplacement.csv")
```

```python
data.isna().sum()
```

```python
data
```

Univarite Analaysis

```python
sns.distplot(data['hsc_p'],hist=True,kde=True,kde_kws={'color':'red'})
# Here hist is histogram
# distplot is used only in univariate can give only one  columns  of the dataset
# Blue colour is histogram and red colour is bell shape curve 
# Kde is kernal density function
# Kde_kws is to make bell shape colour
# The information can get from the graph - the most of the student got 60 marks and the range of mark 60 - 80 only almost of student got mark
```

Bivariate

```python
sns.jointplot(x='mba_p',y='specialisation',data=data)
# The jointplot displays a bivariate analysis.
# It uses numerical values for both the X and Y axes based on columns.
# The jointplot can depict both histograms and scatter plots.
# In the Mkt&Fin specialization, most students scored marks ranging from 55 to 75.
# In the Mkt&HR specialization, marks primarily fell between 55 and 65.
# Mkt&Fin scored well then Mkt&HR
```

```python
sns.jointplot(x='mba_p',y='etest_p',kind='hex',data=data)
# In this hexbin plot, darker shades of blue indicate that more data points where individuals have similar values for the variables 
# The lighter shades of blue indicate that fewer data points have similar scores individuals have the same or similar values for the variables 
# Only fews of the mba_p student scored 60 marks and in etest_p also scored 60 marks.
# The average of the student scored between 55 to 60 marks in mba_p and 55 to 70 marks scored in etest_p
```

```python
sns.jointplot(x='mba_p',y='salary',kind='kde',data=data)
# The student scored in 55 to 80 in mba_p getting average salary of 200000 to 60000
# Most of the student scored 50 to 75 marks getting lesse then 200000 salary.
```

Multivariate

```python
sns.pairplot(data=data,hue='gender',diag_kind='kde',palette='husl',kind='scatter')
# A pairplot creates a grid of subplots where each subplot is a scatter plot (or another type of plot) 
# representing the relationship between a pair of variables. 
# It also includes histograms (or KDE plots) on the diagonals to show the distribution of each variable.
```

Categorical plot

```python
sns.swarmplot(x='ssc_b',y='ssc_p',hue='gender',data=data)
# The relationships between categorical and numerical variables. 
# A swarmplot displays individual data points on a categorical axis, with each point represented as a small marker. 
# It provides a clear view of the distribution of data within categories and helps to show the density of data points.
# The central school of 10th male students scored most of them got 50 to 75 then others school students
```

```python
sns.stripplot(x='ssc_b',y='ssc_p',data=data,hue='gender',jitter=True)
# A stripplot displays individual data points for a categorical variable on a continuous numerical axis. 
# It essentially plots each data point along the y-axis (numerical variable) for each category on the x-axis, 
# often with some jitter to prevent overlap and improve visibility.
# The central school of 10th male students scored most of them got 60 to 70 then others school students
```

Distribution Observation

```python
# Violinplot
sns.violinplot(x='gender',y='ssc_p',data=data)
# In violinplot the male scored 60 to 75 marks in ssc_p and the average is 65.
# Female scored 65 to 80 marks in ssc_p and the average  is 75.
```

```python
# Boxplot
sns.boxplot(x='ssc_b',y='ssc_p',data=data)
# Boxplot is quartile concept the box is starting is Q1 and middle line is Q2 and the third line is Q3 and last line  is Q4 
# The other student who score 25% is 70 and the central student 25% score is 65
```

```python
# Barplot
sns.barplot(x='ssc_b',y='ssc_p',hue="gender",data=data,estimator=np.std)
# barplot typically displays the mean values.
# In this plot displays the std.
# The std.deva of the other students in ssc_p is 11
# The std.deva of the central students in ssc_p is 10
```

```python
# Countplot
sns.countplot(x='ssc_b',hue='gender',data=data)
# Each bar represents the number of instances for each category
# A count plot specifically displays the total count of occurrences for each category within a categorical variable, not for numerical variables.
# The total count of the other male students is 65 lower then the central male student is 70
# The femal student in other and central preform low then male 
```

```python
# Pointplot
sns.pointplot(x='ssc_b',y='ssc_p',hue='gender',data=data)
# In pointplot the other female student mean value is 70 and male student mean value is 65
# The central female students mean value is 68 and male student mean value is 66
```

```python
# Boxplot
sns.boxplot(data=data,orient='h')
```

```python
# Catplot
sns.catplot(x='ssc_b',y='ssc_p',hue='gender',kind='violin',data=data)
# Factorplot is replaced as catplot
# In X axis is catgorical value and Y axis is numbercal value
```

```python
# Regplot 
sns.regplot(x='salary',y='ssc_p',data=data)
# In X and Y axcis is numberical
```

```python
# Lmplot
sns.lmplot(x='salary',y='ssc_p',data=data)
```

```python
# Lmplot with Query
df=data.query('salary > 400000')
sns.lmplot(x='salary',y='ssc_p',data=df)
# In X and Y axcis is numberical
# query is applying condition in lmplot
```
